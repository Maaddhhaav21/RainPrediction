```python
# define numerical & categorical columns
numeric_features = [feature for feature in df.columns if df[feature].dtype != 'O']
categorical_features = [feature for feature in df.columns if df[feature].dtype == 'O']

# print columns
print('We have {} numerical features : {}'.format(len(numeric_features), numeric_features))
print('\nWe have {} categorical features : {}'.format(len(categorical_features), categorical_features))
```

Multivariate analysis using pieplot

```python
plt.rcParams['figure.figsize'] = (30, 12)

# -------------------------------
# Rain Distribution
# -------------------------------

plt.subplot(1,5,1)

size = df['rain_tomorrow'].value_counts()
labels = ['No Rain','Rain']
colors = ['green','blue']

plt.pie(size, colors=colors, labels=labels, autopct='%1.1f%%')
plt.title('Rain Tomorrow')
plt.axis('off')


# -------------------------------
# Location Distribution
# -------------------------------

plt.subplot(1,5,2)

size = df['location'].value_counts()
labels = df['location'].value_counts().index
colors = sns.color_palette('pastel')

plt.pie(size, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Location')
plt.axis('off')
# -------------------------------
# Temperature Distribution
# -------------------------------

plt.subplot(1,5,3)

sns.histplot(data=df, x='temperature', kde=True, color='orange')
plt.title('Temperature Distribution')


# -------------------------------
# Humidity Distribution
# -------------------------------

plt.subplot(1,5,4)

sns.histplot(data=df, x='humidity', kde=True, color='blue')
plt.title('Humidity Distribution')


# -------------------------------
# Cloud Cover Distribution
# -------------------------------

plt.subplot(1,5,5)

sns.histplot(data=df, x='cloud_cover', kde=True, color='purple')
plt.title('Cloud Cover Distribution')


plt.tight_layout()
plt.show()
```
